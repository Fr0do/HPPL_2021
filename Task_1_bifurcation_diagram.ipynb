```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def map_fn(x, r):
    return r * x * (1 - x)
```

```python
def evolution(r=1, n=200, m=200):
    x = []
    x_next = np.random.random()
    for i in range(n + m):
        x_next = map_fn(x_next, r)
        if i > n:
            x.append(x_next)
    return x
```

```python
def plot_evolution(r=1, n=200, m=200):
    x = evolution(r, n, m)
    plt.ylabel('x_n')
    plt.xlabel('n iterations')
    plt.plot(x)
```

```python
plt.figure(figsize=(20,10))
rs = np.linspace(0, 3., 11)
for r in rs:
    plot_evolution(r)
plt.legend(rs.round(3))
plt.show()
```

```python
plt.figure(figsize=(20,10))
rs = np.linspace(3., 3.1, 11)
for r in rs:
    plot_evolution(r)
plt.legend(rs.round(3))
plt.show()
```

```python
plt.figure(figsize=(20, 10))
rs = np.linspace(0, 4., 500)

result = []
for r in rs:
    result.append(evolution(r))

plt.xlabel('r')
plt.ylabel('x_n')
plt.plot(rs, result, 'b.', markersize = 0.1)
plt.show()
```
