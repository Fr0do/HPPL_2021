```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def logistic_map(x, r):
    return r * x * (1 - x)

def exp_map(x, r):
    return x * np.exp(r * (1 - x))

def sin_map(x, r):
    return r * np.sin(np.pi * x)
```

```python
def evolution(r=1, n=200, m=200, map_fn=logistic_map):
    x = []
    x_next = np.random.random()
    for i in range(n + m):
        x_next = map_fn(x_next, r)
        if i > n:
            x.append(x_next)
    return x
```

```python
def plot_evolution(r=1, n=200, m=200, map_fn=logistic_map):
    x = evolution(r, n, m, map_fn)
    plt.ylabel('x_n')
    plt.xlabel('n iterations')
    plt.plot(x)
```

$r \in [0, 1]$: population dies out.

```python
plt.figure(figsize=(20,10))
rs = np.linspace(0., 1., 15)
for r in rs:
    plot_evolution(r, 0, 120)
plt.legend(rs.round(3))
plt.show()
```

$r \in (1, 2]$: population smoothly converges to a stationary value $\frac{r-1}{r}$.

```python
plt.figure(figsize=(20,10))
rs = np.linspace(1., 2., 15)
for r in rs:
    plot_evolution(r, 0, 120)
plt.legend(rs.round(3))
plt.show()
```

$r \in (2, 3]$: population converges to a stationary value $\frac{r-1}{r}$with some oscillations.

```python
plt.figure(figsize=(20,10))
rs = np.linspace(2., 3., 15)
for r in rs:
    plot_evolution(r, 0, 120)
plt.legend(rs.round(3))
plt.show()
```

$r \in (3, 1+\sqrt 6)$: population converges to a stationary value $\frac{r-1}{r}$ oscillating between 2 values.

```python
plt.figure(figsize=(20,10))
rs = np.linspace(3, 1 + np.sqrt(6), 3)
for r in rs:
    plot_evolution(r, 0, 120)
plt.legend(rs.round(3))
plt.show()
```

$r \in [1+\sqrt 6, 4]$: population converges to a stationary value $\frac{r-1}{r}$ oscillating between 4 values, but closer to 4 there are no meaningfill patterns - complete chaos.

```python
plt.figure(figsize=(20,10))
rs = np.linspace(1 + np.sqrt(6), 4, 3)
for r in rs:
    plot_evolution(r, 0, 120)
plt.legend(rs.round(3))
plt.show()
```

$r > 4$: population diverges to $\infty$.

```python
plt.figure(figsize=(20,10))
rs = np.linspace(4, 4.1, 3)
for r in rs:
    plot_evolution(r, 0, 120)
plt.legend(rs.round(3))
plt.show()
```

```python
plt.figure(figsize=(20, 10))
rs = np.linspace(0, 4., 500)

log_res = []
exp_res = []
sin_res = []
for r in rs:
    log_res.append(evolution(r))
    sin_res.append(evolution(r, map_fn=sin_map))
    exp_res.append(evolution(r, map_fn=exp_map))

plt.xlabel('r')
plt.ylabel('x_n')

plt_log = plt.plot(rs, log_res, 'r.', markersize = 0.1)
plt_sin = plt.plot(rs, sin_res, 'g.', markersize = 0.1)
plt_exp = plt.plot(rs, exp_res, 'b.', markersize = 0.1)
plt.legend(labels=['logistic', 'sin', 'exp'], labelcolor=['r', 'g', 'b'])
plt.show()
```

```python
plt.figure(figsize=(20, 10))
rs = np.linspace(3., 4., 500)

result = []
for r in rs:
    result.append(evolution(r))

plt.xlabel('r')
plt.ylabel('x_n')
plt.plot(rs, result, 'b.', markersize = 0.1)
plt.show()
```
