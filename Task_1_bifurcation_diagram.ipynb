```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
def logistic_map(x, r):
    return r * x * (1 - x)

def exp_map(x, r):
    return x * np.exp(r * (1 - x))

def sin_map(x, r):
    return r * np.sin(np.pi * x)
```

```python
def evolution(r=1, n=200, m=200, map_fn=logistic_map):
    x = []
    x_next = np.random.random()
    for i in range(n + m):
        x_next = map_fn(x_next, r)
        if i > n:
            x.append(x_next)
    return x
```

```python
def plot_evolution(r=1, n=200, m=200, map_fn=logistic_map):
    x = evolution(r, n, m, map_fn)
    plt.ylabel('x_n')
    plt.xlabel('n iterations')
    plt.plot(x)
```

```python
plt.figure(figsize=(20,10))
rs = np.linspace(0, 3., 11)
for r in rs:
    plot_evolution(r, map_fn=exp_map)
plt.legend(rs.round(3))
plt.show()
```

```python
plt.figure(figsize=(20,10))
rs = np.linspace(3., 3.1, 11)
for r in rs:
    plot_evolution(r)
plt.legend(rs.round(3))
plt.show()
```

```python
plt.figure(figsize=(20, 10))
rs = np.linspace(0, 4., 500)

log_res = []
exp_res = []
sin_res = []
for r in rs:
    log_res.append(evolution(r))
    sin_res.append(evolution(r, map_fn=sin_map))
    exp_res.append(evolution(r, map_fn=exp_map))

plt.xlabel('r')
plt.ylabel('x_n')

plt_log = plt.plot(rs, log_res, 'r.', markersize = 0.1)
plt_sin = plt.plot(rs, sin_res, 'g.', markersize = 0.1)
plt_exp = plt.plot(rs, exp_res, 'b.', markersize = 0.1)
plt.legend(labels=['logistic', 'sin', 'exp'], labelcolor=['r', 'g', 'b'])
plt.show()
```
